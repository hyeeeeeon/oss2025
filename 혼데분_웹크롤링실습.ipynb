```python
# isbn으로 상세 페이지 가져오기 + format() 사용
import requests as rq
isbn = 	9791162540633
url = 'https://www.yes24.com/product/search?domain=BOOK&query={}'
r = rq.get(url.format(isbn))

print(r.text)
```

```python
# a 태그에서 class 명 가지고 href 주소 추출하기 + find() 이용

from bs4 import BeautifulSoup as bs

soup = bs(r.text, 'html.parser')
prd_link = soup.find('a', attrs={'class': 'gd_name'})

print(prd_link)
```

```python
print(prd_link['href'])
```

```python
# 그릿 상세 페이지 가져오기
url = 'http://www.yes24.com' + prd_link['href']
r = rq.get(url)

print(r.text)
```

```python
# 쪽수 정보 가져오기
soup = bs(r.text, 'html.parser')
prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
print(prd_detail)
```

```python
# 테이블 태그 리스트로 가져오기 => find_all()
prd_tr = prd_detail.find_all('tr')
print(prd_tr)
```

```python
# 태그 안에 텍스트 가져오기 => get_text()
for tr in prd_tr:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
        page_pd = tr.find('td').get_text()
        break

print(page_pd)
```

```python
# 페이지만 가져오기
print(page_pd.split()[0])
```

```python
# 위 내용 활용 전체 도서 쪽 수 구하기
def get_page(isbn):
    #yse24 검색 페이지 url
    url = 'https://www.yes24.com/product/search?domain=BOOK&query={}'
    r = rq.get(url.format(isbn))
    soup = bs(r.text, 'html.parser')
    prd_info = soup.find('a', attrs={'class': 'gd_name'})
    if prd_info == None:
        return 'None'

    # 도서 상세 페이지 가져오기
    url = 'http://www.yes24.com' + prd_info['href']
    r = rq.get(url)
    soup = bs(r.text, 'html.parser')
    # 품목 정보 <div> 선택
    prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
    # 테이블 <tr> 태그 가져오기
    prd_tr = prd_detail.find_all('tr')
    # 쪽수가 있는 <th> 찾고, 거기 <td> 값 반환
    for tr in prd_tr:
        if tr.find('th').get_text() == '쪽수, 무게, 크기':
            return tr.find('td').get_text().split()[0]

    return ' '
```

```python
get_page(9791162540633)
```

```python
top10_books = books.head(10)
top10_books
```

```python
def get_page2(row):
    isbn = row['ISBN']
    return get_page(isbn)
```

```python
page_count = top10_books.apply(get_page2, axis=1)       # aixs = 0 은 열 기준
print(page_count)
```

```python
page_count.name = '쪽수'
print(page_count)
```

```python
top10_page = pd.merge(top10_books, page_count, left_index=True, right_index=True)
top10_page
```
